```python
import pandas as pd
import yfinance as yf
import ta
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, confusion_matrix, classification_report
import numpy as np

# --- Step 1 & 2: Data Loading and Feature Engineering ---
# Load historical data for AAPL from 2020-01-01 to 2023-01-01
print("Loading historical data...")
data = yf.download("AAPL", start="2020-01-01", end="2023-01-01")

# Rename columns to standard names based on their position (assuming 10 columns initially)
# We'll check the actual number of columns and adjust if necessary
if data.shape[1] == 10:
    data.columns = ['Open', 'High', 'Low', 'Close', 'Volume', 'Adj Close', 'Dividends', 'Stock Splits', 'Capital Gains', 'Ignore']
elif data.shape[1] == 7: # Sometimes yfinance returns 7 columns
     data.columns = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume', 'Ignore']
elif data.shape[1] == 6: # Sometimes yfinance returns 6 columns (older versions)
     data.columns = ['Open', 'High', 'Low', 'Close', 'Volume', 'Adj Close']
elif data.shape[1] == 5: # Based on the future_data experience
     data.columns = ['Open', 'High', 'Low', 'Close', 'Volume']
else:
    print(f"Unexpected number of columns: {data.shape[1]}. Columns: {data.columns}")
    # Attempt to flatten if it's a MultiIndex, although previous inspection showed it wasn't
    if isinstance(data.columns, pd.MultiIndex):
         data.columns = data.columns.get_level_values(-1)
         print("Attempted to flatten MultiIndex.")
    # Fallback: use positional renaming if flattening didn't work or wasn't needed
    if 'Close' not in data.columns and data.shape[1] >= 4:
         print("Renaming columns based on position as 'Close' is not found.")
         col_names = ['Open', 'High', 'Low', 'Close']
         # Add 'Volume' if available
         if data.shape[1] >= 5:
             col_names.append('Volume')
         # Add other columns if available
         if data.shape[1] > 5:
             col_names.extend([f'Col{i}' for i in range(5, data.shape[1])])
         data.columns = col_names


print("Feature engineering...")
# --- Moving Averages ---
data['SMA_5'] = data['Close'].rolling(window=5).mean()
data['SMA_10'] = data['Close'].rolling(window=10).mean()
data['SMA_50'] = data['Close'].rolling(window=50).mean()
data['EMA_10'] = data['Close'].ewm(span=10, adjust=False).mean()
data['EMA_50'] = data['Close'].ewm(span=50, adjust=False).mean()

# --- Technical Indicators (Example: RSI) ---
rsi = ta.momentum.RSIIndicator(close=data['Close'], window=14)
data['RSI'] = rsi.rsi()

# --- Technical Indicators (Example: MACD) ---
macd = ta.trend.MACD(close=data['Close'])
data['MACD'] = macd.macd()
data['MACD_Signal'] = macd.macd_signal()

# --- Volatility Measures: ATR ---
atr = ta.volatility.AverageTrueRange(high=data['High'], low=data['Low'], close=data['Close'], window=14)
data['ATR'] = atr.average_true_range()


# Drop rows with NaN values created by feature engineering
data.dropna(inplace=True)

print("Shape of data after feature engineering and dropna():", data.shape)
if data.empty:
    print("Data is empty after feature engineering. Cannot proceed with modeling.")
else:
    # --- Step 3: Model Building ---
    print("\nBuilding and training models...")

    # --- Prepare data for Regression Model (Predicting Close Price) ---
    data_reg = data.copy()
    data_reg['Target_Close'] = data_reg['Close'].shift(-1) # Predict the next day's closing price
    data_reg.dropna(inplace=True) # Drop the last row with NaN target

    if data_reg.empty:
         print("Data for regression is empty after creating target. Cannot train regression model.")
    else:
        features = ['Open', 'High', 'Low', 'Close', 'Volume', 'SMA_5', 'SMA_10', 'SMA_50', 'EMA_10', 'EMA_50', 'RSI', 'MACD', 'MACD_Signal', 'ATR']
        # Ensure features exist in the DataFrame
        features_reg = [f for f in features if f in data_reg.columns]
        X_reg = data_reg[features_reg]
        y_reg = data_reg['Target_Close']


        # Split data into training and testing sets (70/30 ratio)
        if len(X_reg) > 1: # Ensure there's enough data to split
            X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.3, random_state=42, shuffle=False) # shuffle=False for time series data

            # --- Train Linear Regression Model ---
            reg_model = LinearRegression()
            reg_model.fit(X_train_reg, y_train_reg)

            # Make predictions
            y_pred_reg = reg_model.predict(X_test_reg)

            # Evaluate Regression Model
            mse = mean_squared_error(y_test_reg, y_pred_reg)
            r2 = r2_score(y_test_reg, y_pred_reg)

            print("--- Linear Regression Model Performance (Predicting Close Price) ---")
            print(f"Mean Squared Error (MSE): {mse:.2f}")
            print(f"R-squared (R²): {r2:.2f}")

            # Store the trained regression model
            trained_reg_model = reg_model
        else:
            print("Not enough data for train-test split for regression.")


    # --- Prepare data for Random Forest Classification Model (Predicting Trend) ---
    data_clf = data.copy()
    data_clf['Price_Increase'] = (data_clf['Close'].shift(-1) > data_clf['Close']).astype(int)
    data_clf.dropna(inplace=True) # Drop the last row with NaN target


    if data_clf.empty:
        print("\nData for classification is empty after creating target. Cannot train classification model.")
    else:
        features_clf = [f for f in features if f in data_clf.columns] # Use the same features, ensure they exist
        X_clf = data_clf[features_clf]
        y_clf = data_clf['Price_Increase']

        # Split data into training and testing sets (70/30 ratio)
        if len(X_clf) > 1: # Ensure there's enough data to split
            X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X_clf, y_clf, test_size=0.3, random_state=42, shuffle=False) # shuffle=False for time series data


            # --- Train Random Forest Classification Model ---
            clf_model = RandomForestClassifier(n_estimators=100, random_state=42)
            clf_model.fit(X_train_clf, y_train_clf)

            # Make predictions
            y_pred_clf = clf_model.predict(X_test_clf)

            # Evaluate Classification Model
            accuracy = accuracy_score(y_test_clf, y_pred_clf)
            conf_matrix = confusion_matrix(y_test_clf, y_pred_clf)
            class_report = classification_report(y_test_clf, y_pred_clf)

            print("\n--- Random Forest Classification Model Performance (Predicting Trend) ---")
            print(f"Accuracy: {accuracy:.2f}")
            print("\nConfusion Matrix:")
            print(conf_matrix)
            print("\nClassification Report:")
            print(class_report)

            # Store the trained classification model
            trained_clf_model = clf_model
        else:
             print("Not enough data for train-test split for classification.")
```

- **Action**: Buy stocks when the model predicts an increase (class 1).

- **Focus**: Take advantage of the model's high recall for no price increase predictions (class 0).

#### Hybrid Strategy:

- **Combine**: Use both conservative and contrarian strategies based on confidence thresholds.
- **Action**: Set a higher threshold for buying (only buy when the predicted probability of price increase is significantly high) and a lower threshold for selling or staying out of the market.
- **Risk Management**: Adjust thresholds dynamically based on recent model performance and market volatility.
